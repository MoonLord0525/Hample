###  auROC performance comparison of Hample with several state-of-the-art TFBS prediction models

```python
import pandas as pd
data = pd.read_excel("./result/compare_roc.xlsx", sheet_name="Sheet1")
data
```

```python
order = ['Hample(0.954)', "DeepBind(0.855)", "DeeperBind(0.862)", "DeepSite(0.877)",
             "HOCNN(0.877)", "DLBSS(0.865)", "CRPTS(0.859)", "CNN_TF(0.891)"]
data_order = pd.DataFrame(columns=data.columns)
data_grp = data.groupby("model")
for o in order:
    frame = data_grp.get_group(o)
    data_order = data_order.append(frame)
data_order.reset_index(inplace=True, drop=True)
data_order
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 4))

plt.rc('font', family='Times New Roman', size=13, weight='bold')
plt.rc('axes', axisbelow=True)
fontdict_label = {'family': "Times New Roman", 'size': 13, 'weight': "bold"}

colors = ["#5470c6", "#91cc75", "#fac858", "#ee6666", "#73c0de", "#3ba272", "#fc8452", "#9a60b4"]

ax = sns.barplot(data=data_order, x='tf', y='roc_auc', hue='model',
                      palette=colors, edgecolor='black', zorder=10)
    # set width
    # w = 0.15
    # for bar in ax.patches:
    #     x = bar.get_x()
    #     width = bar.get_width()
    #     centre = x + width / 2.
    #
    #     bar.set_x(centre - w / 2.)
    #     bar.set_width(w)
ax.legend(loc=2, ncol=4,
              frameon=False, columnspacing=1.5, handlelength=0.7)

ax.set_ylabel("auROC", fontdict_label)
ax.set_xlabel("TFs", fontdict_label)
ax.set_ylim((0.7, 1.1))
ax.grid(axis='y', linestyle='--', alpha=0.8, zorder=0, linewidth=0.8, color='gray')
ax.spines[['right', 'bottom', 'right', 'left']].set_color('black')
ax.spines[['right', 'bottom', 'right', 'left']].set_linewidth('1.5')
```

###  auROC performance of Hample on cell-type shared and cell-type-specific TFBSs in the five cell types

```python
import pandas as pd

specific = pd.read_excel("./Cell-types/Specific.xlsx", sheet_name="Sheet1", index_col=0)
specific = specific.drop("USF2", axis=0)
specific
```

```python
share = pd.read_excel("./Cell-types/Share.xlsx", sheet_name="Sheet1", index_col=0)
share = share.drop("USF2", axis=0)
share
```

```python
from Utils.util_function import get_cells

data_fig = pd.DataFrame(columns=['cell', 'hueType', 'sensitivity'])
cells = get_cells()
data_tmp = pd.DataFrame(columns=['cell', 'hueType', 'sensitivity'])
for cell in cells:
    data_tmp['sensitivity'] = specific[cell].tolist() + share[cell].tolist()
    data_tmp['cell'] = [cell] * 14
    data_tmp['hueType'] = ["Cell-type specific TFBS"] * 7 + ["Cell-type shared TFBS"]*7
    data_fig = data_fig.append(data_tmp)
data_fig
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family='Times New Roman', size=13, weight='bold')
plt.rc('axes', axisbelow=True)

fontdict_label = {'family': "Times New Roman", 'size': 13, 'weight': "bold"}
colors = ["#FACDAD", "#81B09B"]

ax = sns.boxplot(
    data=data_fig, x="cell", y='sensitivity', hue='hueType',
    fliersize=2.5,
    showmeans=True, meanline=True,
    meanprops=dict(color='black', linestyle='-', linewidth=1.5),
    medianprops=dict(linewidth=0),
    palette=colors,)
ax.spines[['right', 'bottom', 'right', 'left']].set_color('black')
ax.spines[['right', 'bottom', 'right', 'left']].set_linewidth('1.5')
ax.grid(axis='y', linestyle='--', alpha=0.8, zorder=0, linewidth=0.8, color='gray')
ax.set_ylabel("auROC", fontdict_label)
ax.set_xlabel("Cell-types", fontdict_label)
plt.ylabel("Sensitivity", fontdict_label)
plt.xlabel("Cell-types", fontdict_label)
plt.legend(frameon=False, title=None)
plt.show()
```

### Pearson correlation of cell-type-shared and cell-type-specific TFBSs in the five cell types

```python
import pandas as pd
pear_share = pd.read_csv("./pearsonr/sharedpearsonr.csv", index_col=0)
for i in pear_share.index:
    if i.startswith("USF2") or i.endswith("USF2"):
        pear_share = pear_share.drop(i, axis=0)
pear_share
```

### Pearson correlation of binding sites between each pair-wise TF

```python
import pandas as pd
data = pd.read_csv("./pearsonr/pearsonr.csv", index_col=0)
data
```

```python
from Utils.util_function import get_tfs
import numpy as np

matrix = np.zeros(shape=(8, 8))
TFS = get_tfs()
pear_index = 0
pear_mean = np.mean(data, 1)
for i in range(8):
    for j in range(i + 1, 8):
        matrix[i, j] = pear_mean[pear_index]
        matrix[j, i] = matrix[i, j]
        pear_index += 1
matrix = pd.DataFrame(matrix)
matrix.columns = TFS + ["USF2"]
matrix.index = TFS + ['USF2']
matrix = matrix.drop('USF2', axis=0)
matrix = matrix.drop("USF2", axis=1)
matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Times New Roman', size=13, weight='bold')
plt.rc('axes', axisbelow=True)

fontdict_label = {'family': "Times New Roman", 'size': 13, 'weight': "bold"}
ax = sns.heatmap(matrix, cmap='OrRd', linewidths=0.5,linecolor='white',vmin=0, vmax=1.0, square=True)
ax.set_title("Pearson Correlation = 0.786", fontdict_label)

# plt.savefig("pear_heatmap.jpg",  bbox_inches='tight', dpi=300)
plt.show()
```
